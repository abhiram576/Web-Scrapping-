# as a function

```python
urls=[]
movie_name=[]
next_page_url="http://www.imfdb.org/wiki/Category:Movie"
while True:
    response=requests.get(next_page_url)
    soup=BeautifulSoup(response.text,"html.parser")
    movie_pages=soup.find("div",attrs={"id":"mw-pages"})
    # for movie names and url
    for i in movie_pages("li"):
        movie_name.append(i.text)
        url="http://www.imfdb.org"+i.a["href"]
        urls.append(url)
    next_page=movie_pages("a")[-1]
    if next_page.text=="next 200":
        next_page_url="http://www.imfdb.org"+next_page["href"]
        print(next_page_url)
    else:
        break
    
        
```
